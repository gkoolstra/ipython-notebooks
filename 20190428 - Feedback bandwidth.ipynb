```python
Q = 38E3
f0 = 6.05E9
kappa_tot = f0 / Q
kappa1 = kappa_tot / 2.
kappa2 = kappa1

f = np.logspace(0, 6, 1E3)

mixed_down = np.abs(np.sqrt(kappa1 * kappa2) / (1j * f + (kappa1 + kappa2) / 2.))

plt.figure()
plt.plot(f, mixed_down)
plt.xscale('log')
plt.yscale('log')
plt.ylim(1E-2, 1.05)

plt.figure()
plt.plot(f, mixed_down)
plt.plot(-f, mixed_down)
plt.ylim(1E-2, 1.05)
```
