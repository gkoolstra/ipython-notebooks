```python
V = np.linspace(0, 5)
eps0 = 8.85E-12
epsHe = 1.056
d = 1.1E-6
sigma = 3.78E-4
w = 3.5E-6
```

```python
def b(V):
    return epsHe * eps0 * V**2 / d**2 * (8 * sigma/w**2 - 2*V**2 * epsHe * eps0 / d**3)**(-1)
```

```python
plt.figure()
common.configure_axes(13)
plt.plot(V, (d - b(V)) * 1E6)
plt.hlines(d * 1E6, 0, np.max(V), linestyles='--', color='gray', lw=2)
# plt.yscale('log')
plt.xlabel("Resonator bias voltage (V)")
plt.ylabel("Helium depth in center of channel $d-b$")
plt.ylim(0, 1.2)
```
